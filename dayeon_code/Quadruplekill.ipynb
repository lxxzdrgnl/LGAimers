```python
# ê¸°ì¡´ "ì‹œìˆ  ìœ í˜•" ì»¬ëŸ¼ ì‚­ì œ
df_IVF.drop(columns=["ì‹œìˆ  ìœ í˜•"], inplace=True, errors="ignore")
df_DI.drop(columns=["ì‹œìˆ  ìœ í˜•"], inplace=True, errors="ignore")

# ìƒˆë¡œìš´ "ì‹œìˆ  ìœ í˜•" ì»¬ëŸ¼ ì¶”ê°€
df_IVF["ì‹œìˆ  ìœ í˜•"] = 1
df_DI["ì‹œìˆ  ìœ í˜•"] = 0
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MultiLabelBinarizer

# 1ï¸âƒ£ **"ì•Œ ìˆ˜ ì—†ìŒ"ì„ 'Unknown'ìœ¼ë¡œ ì²˜ë¦¬í•  ì»¬ëŸ¼ ë¦¬ìŠ¤íŠ¸**
columns_to_replace = ["ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´", "ë‚œì ì¶œì²˜", "ë‚œì ê¸°ì¦ì ë‚˜ì´", "ì •ì ê¸°ì¦ì ë‚˜ì´", "ë°°ë€ ìœ ë„ ìœ í˜•", "ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ "]

for col in columns_to_replace:
    if col in df_IVF.columns:
        df_IVF[col] = df_IVF[col].replace("ì•Œ ìˆ˜ ì—†ìŒ", "Unknown")  # âœ… "-1" ëŒ€ì‹  "Unknown" ì‚¬ìš©

# 2ï¸âƒ£ **NaN ê°’ ë¨¼ì € ì²˜ë¦¬ (ì´ì œ ê²°ì¸¡ì¹˜ ì•ˆ ìƒê¹€!)**
df_IVF.fillna("Unknown", inplace=True)  # âœ… ëª¨ë“  NaNì„ "Unknown"ìœ¼ë¡œ ë³€í™˜

# 3ï¸âƒ£ **Label Encoding (ë¬¸ìì—´ ë³€í™˜ í›„ ì§„í–‰)**
label_encode_columns = [
    "ì‹œìˆ  ì‹œê¸° ì½”ë“œ", "ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´", "ë°°ë€ ìœ ë„ ìœ í˜•", "ì´ ì‹œìˆ  íšŸìˆ˜", "í´ë¦¬ë‹‰ ë‚´ ì´ ì‹œìˆ  íšŸìˆ˜",
    "IVF ì‹œìˆ  íšŸìˆ˜", "DI ì‹œìˆ  íšŸìˆ˜", "ì´ ì„ì‹  íšŸìˆ˜", "IVF ì„ì‹  íšŸìˆ˜", "DI ì„ì‹  íšŸìˆ˜",
    "ì´ ì¶œì‚° íšŸìˆ˜", "IVF ì¶œì‚° íšŸìˆ˜", "DI ì¶œì‚° íšŸìˆ˜", "ë‚œì ì¶œì²˜", "ì •ì ì¶œì²˜",
    "ë‚œì ê¸°ì¦ì ë‚˜ì´", "ì •ì ê¸°ì¦ì ë‚˜ì´"
]

le = LabelEncoder()
for col in label_encode_columns:
    if col in df_IVF.columns:
        df_IVF[col] = df_IVF[col].astype(str)  # âœ… ëª¨ë“  ê°’ì„ ë¬¸ìì—´ë¡œ ë³€í™˜ (NaN ë°©ì§€)
        df_IVF[col] = le.fit_transform(df_IVF[col])  # âœ… LabelEncoder ì ìš©!

# 4ï¸âƒ£ **"ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ " ì›-í•« ì¸ì½”ë”© (NaN ì²˜ë¦¬ í›„ ì§„í–‰)**
if "ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ " in df_IVF.columns:
    df_IVF["ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ "] = df_IVF["ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ "].astype(str).apply(lambda x: x.split(',') if x != "Unknown" else [])
    
    mlb = MultiLabelBinarizer()
    encoded_ë°°ì•„ìƒì„± = mlb.fit_transform(df_IVF["ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ "])
    encoded_ë°°ì•„ìƒì„±_df = pd.DataFrame(encoded_ë°°ì•„ìƒì„±, columns=[f"ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ _{c}" for c in mlb.classes_])
    
    df_IVF.drop("ë°°ì•„ ìƒì„± ì£¼ìš” ì´ìœ ", axis=1, inplace=True)
    df_IVF = pd.concat([df_IVF, encoded_ë°°ì•„ìƒì„±_df], axis=1)

# 5ï¸âƒ£ **Label Encoding ì´í›„ ë‹¤ì‹œ ê²°ì¸¡ì¹˜ê°€ ìƒê¸°ëŠ” ê²½ìš° ëŒ€ë¹„**
df_IVF.fillna(0, inplace=True)  # âœ… ëª¨ë“  NaNì„ 0ìœ¼ë¡œ ë³€í™˜ (ê²°êµ­ ì™„ë²½í•œ ë°ì´í„°!)

# 6ï¸âƒ£ **ë§ˆì§€ë§‰ ê²°ì¸¡ì¹˜ í™•ì¸ (ì´ì œ ì—†ì–´ì•¼ í•¨)**
missing_values_IVF = df_IVF.isnull().sum()
missing_values_IVF = missing_values_IVF[missing_values_IVF > 0]  # ê²°ì¸¡ì¹˜ ìˆëŠ” ì»¬ëŸ¼ë§Œ

print("\nğŸ“Œ IVF ë°ì´í„° ê²°ì¸¡ì¹˜ í™•ì¸:")
if missing_values_IVF.empty:
    print("âœ… ê²°ì¸¡ì¹˜ê°€ ì—†ìŠµë‹ˆë‹¤! ğŸ‰")
else:
    print(missing_values_IVF)
```

```python
###############################3ì—¬ê¸°ì„œë¶€í„° 

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# âœ… IVF & DI ë°ì´í„°ì…‹ ë¦¬ìŠ¤íŠ¸
df_parts = {"IVF": df_IVF, "DI": df_DI}

valid_treatments = ["IVF", "IUI", "FER", "IVI", "BLASTOCYST", "AH", "GIFT", "ICI", "Generic DI", "ICSI", "Unknown"]

# IVF ë° DI ë°ì´í„° ì²˜ë¦¬
for df_name, df_part in df_parts.items():
    if "íŠ¹ì • ì‹œìˆ  ìœ í˜•" in df_part.columns:
        # ì‹œìˆ  ìœ í˜•ì„ ìœ íš¨í•œ ê°’ì— ë§ì¶° ë³€í™˜ (0 ë˜ëŠ” 1ë¡œ í‘œì‹œ)
        for treatment in valid_treatments:
            df_part[f"íŠ¹ì • ì‹œìˆ  - {treatment}"] = (df_part["íŠ¹ì • ì‹œìˆ  ìœ í˜•"] == treatment).astype(int)
        
        # MIXED ê°’ ì²˜ë¦¬: ìœ íš¨í•˜ì§€ ì•Šì€ ì‹œìˆ  ìœ í˜•ì€ "MIXED"ë¡œ ì²˜ë¦¬
        df_part["íŠ¹ì • ì‹œìˆ  - MIXED"] = 0
        mixed_values = df_part[~df_part["íŠ¹ì • ì‹œìˆ  ìœ í˜•"].isin(valid_treatments)]["íŠ¹ì • ì‹œìˆ  ìœ í˜•"]
        
        if not mixed_values.empty:
            # mixed_valuesì˜ ë°ì´í„°ë¥¼ ë¬¸ìì—´ë¡œ ë³€í™˜í•˜ì—¬ ë ˆì´ë¸” ì¸ì½”ë”©
            mixed_values = mixed_values.astype(str)
            train_encoded = LabelEncoder().fit_transform(mixed_values) + 1
            df_part.loc[~df_part["íŠ¹ì • ì‹œìˆ  ìœ í˜•"].isin(valid_treatments), "íŠ¹ì • ì‹œìˆ  - MIXED"] = train_encoded
        
        # "íŠ¹ì • ì‹œìˆ  ìœ í˜•" ì»¬ëŸ¼ ì œê±°
        df_part.drop(columns=["íŠ¹ì • ì‹œìˆ  ìœ í˜•"], inplace=True)

# ì²˜ë¦¬ í›„ ë‹¤ì‹œ í™•ì¸
for df_name, df_part in df_parts.items():
    if "íŠ¹ì • ì‹œìˆ  ìœ í˜•" not in df_part.columns:
        print(f"âœ… '{df_name}' ë°ì´í„°ì—ì„œ 'íŠ¹ì • ì‹œìˆ  ìœ í˜•' ì»¬ëŸ¼ì´ ì •ìƒì ìœ¼ë¡œ ì œê±°ë˜ì—ˆìŠµë‹ˆë‹¤.")
    else:
        print(f"âš ï¸ '{df_name}' ë°ì´í„°ì—ì„œ 'íŠ¹ì • ì‹œìˆ  ìœ í˜•' ì»¬ëŸ¼ì´ ì—¬ì „íˆ ì¡´ì¬í•©ë‹ˆë‹¤.")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# âœ… IVF & DI ë°ì´í„°ì…‹ ë¦¬ìŠ¤íŠ¸
df_parts = {"IVF": df_IVF, "DI": df_DI}

# --------------------------------------- "ì•Œ ìˆ˜ ì—†ìŒ"ì„ -1ë¡œ ë³€ê²½ (NaN ë°©ì§€) ---------------------------------------------------
columns_to_replace = ["ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´", "ë‚œì ì¶œì²˜", "ë‚œì ê¸°ì¦ì ë‚˜ì´", "ì •ì ê¸°ì¦ì ë‚˜ì´", "ë°°ë€ ìœ ë„ ìœ í˜•"]

for df_name, df_part in df_parts.items():
    for col in columns_to_replace:
        if col in df_part.columns:
            df_part[col] = df_part[col].astype(str)  # âœ… ëª¨ë“  ê°’ì„ ë¬¸ìì—´ë¡œ ë³€í™˜
            df_part[col] = df_part[col].replace("ì•Œ ìˆ˜ ì—†ìŒ", "Unknown")  # âœ… "ì•Œ ìˆ˜ ì—†ìŒ" â†’ "Unknown"
            df_part[col] = df_part[col].replace("-1", "Unknown")  # âœ… í˜¹ì‹œ ìˆì„ "-1"ë„ ë³€í™˜

# --------------------------------------- Label Encoding (NaN ë°©ì§€) ---------------------------------------------------
label_encode_columns = [
    "ì‹œìˆ  ì‹œê¸° ì½”ë“œ", "ì‹œìˆ  ë‹¹ì‹œ ë‚˜ì´", "ë°°ë€ ìœ ë„ ìœ í˜•", "ì´ ì‹œìˆ  íšŸìˆ˜", "í´ë¦¬ë‹‰ ë‚´ ì´ ì‹œìˆ  íšŸìˆ˜",
    "IVF ì‹œìˆ  íšŸìˆ˜", "DI ì‹œìˆ  íšŸìˆ˜", "ì´ ì„ì‹  íšŸìˆ˜", "IVF ì„ì‹  íšŸìˆ˜", "DI ì„ì‹  íšŸìˆ˜",
    "ì´ ì¶œì‚° íšŸìˆ˜", "IVF ì¶œì‚° íšŸìˆ˜", "DI ì¶œì‚° íšŸìˆ˜", "ë‚œì ì¶œì²˜", "ì •ì ì¶œì²˜",
    "ë‚œì ê¸°ì¦ì ë‚˜ì´", "ì •ì ê¸°ì¦ì ë‚˜ì´"
]

for df_name, df_part in df_parts.items():
    for col in label_encode_columns:
        if col in df_part.columns:
            le = LabelEncoder()
            df_part[col] = df_part[col].astype(str)  # âœ… ëª¨ë“  ê°’ì„ ë¬¸ìì—´ë¡œ ë³€í™˜
            df_part[col] = le.fit_transform(df_part[col])  # âœ… LabelEncoder ì ìš©

# --------------------------------------- Min-Max Scaling (ìˆ˜ì¹˜í˜• ë°ì´í„°) ---------------------------------------------------
for df_name, df_part in df_parts.items():
    non_binary_columns = [col for col in df_part.columns if df_part[col].dtype in ['int64', 'float64'] and df_part[col].nunique() > 2]
    
    if len(non_binary_columns) > 0:
        df_part.fillna(0, inplace=True)  # âœ… NaNì´ ë‚¨ì•„ìˆì„ ê°€ëŠ¥ì„± ì œê±°
        scaler = MinMaxScaler()
        df_part[non_binary_columns] = scaler.fit_transform(df_part[non_binary_columns])

# --------------------------------------- ìµœì¢… ê²€ì¦ (ê²°ì¸¡ì¹˜ ë° ë¬¸ìì—´ í™•ì¸) ---------------------------------------------------
for df_name, df_part in df_parts.items():
    missing_values = df_part.isnull().sum()
    missing_values = missing_values[missing_values > 0]

    if missing_values.empty:
        print(f"\nâœ… {df_name} ë°ì´í„° ì „ì²˜ë¦¬ ì™„ë£Œ! ê²°ì¸¡ì¹˜ ì—†ìŒ âœ…")
    else:
        print(f"\nâš ï¸ {df_name} ë°ì´í„°ì— ë‚¨ì•„ìˆëŠ” ê²°ì¸¡ì¹˜ ë°œê²¬! í™•ì¸ í•„ìš” âš ï¸")
        print(missing_values)

    object_cols = df_part.select_dtypes(include=["object"]).columns.tolist()
    if object_cols:
        print(f"\nğŸ“Œ {df_name} ë°ì´í„°ì—ì„œ ë¬¸ìì—´ ì»¬ëŸ¼ í™•ì¸:")
        print(object_cols)
    else:
        print(f"\nâœ… {df_name} ë°ì´í„°ì— ë¬¸ìì—´ ì»¬ëŸ¼ì´ ì—†ìŠµë‹ˆë‹¤.")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, pointbiserialr, chi2_contingency

def cramers_v(x, y):
    """ì´ì§„í˜• ë³€ìˆ˜ ê°„ì˜ ìƒê´€ê´€ê³„ë¥¼ ê³„ì‚°í•˜ëŠ” CramÃ©r's V (ì˜¤ë¥˜ ë°©ì§€ ì¶”ê°€)"""
    confusion_matrix = pd.crosstab(x, y)

    if confusion_matrix.shape[0] == 1 or confusion_matrix.shape[1] == 1:
        return 0  # ì°¨ì›ì´ 1x1ì´ë©´ CramÃ©r's VëŠ” 0ìœ¼ë¡œ ì„¤ì •
    
    chi2, _, _, _ = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    
    min_dim = min(confusion_matrix.shape) - 1
    if min_dim == 0 or n == 0:
        return 0  # 0ìœ¼ë¡œ ë‚˜ëˆ„ëŠ” ê²½ìš° ë°©ì§€
    
    return np.sqrt(chi2 / (n * min_dim))

def calculate_correlations(df, target_variable="ì„ì‹  ì„±ê³µ ì—¬ë¶€"):
    """ë°ì´í„°í”„ë ˆì„ì—ì„œ target_variableê³¼ ìœ ì˜ë¯¸í•œ ìƒê´€ê´€ê³„ë¥¼ ê°–ëŠ” ë³€ìˆ˜ ì¶œë ¥"""
    
    correlations = []

    for col in df.columns:
        if col == target_variable:
            continue
        
        unique_vals = df[col].dropna().unique()
        is_binary = sorted(unique_vals) in [[0, 1], [0], [1]]

        try:
            if is_binary:  # ì´ì§„ ë³€ìˆ˜ ì²˜ë¦¬
                if sorted(df[target_variable].unique()) in [[0, 1], [0], [1]]:  # ì´ì§„í˜• vs ì´ì§„í˜•
                    corr = cramers_v(df[col], df[target_variable])
                    p_val = chi2_contingency(pd.crosstab(df[col], df[target_variable]))[1]
                else:  # ì´ì§„í˜• vs ì—°ì†í˜•
                    corr, p_val = pointbiserialr(df[col], df[target_variable])
            else:  # ì—°ì†í˜• ë³€ìˆ˜ ì²˜ë¦¬
                corr, p_val = pearsonr(df[col].dropna(), df[target_variable].dropna())

            correlations.append((col, corr, p_val))
        
        except Exception as e:
            print(f"âš ï¸ {col} ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

    # ìœ ì˜ìˆ˜ì¤€ 0.05 ì´í•˜ì¸ ë³€ìˆ˜ë§Œ ì •ë ¬í•˜ì—¬ ì¶œë ¥
    significant_vars = sorted([(col, corr, p_val) for col, corr, p_val in correlations if p_val < 0.05], key=lambda x: abs(x[1]), reverse=True)

    return pd.DataFrame(significant_vars, columns=["ë³€ìˆ˜", "ìƒê´€ê³„ìˆ˜", "p-value"])

# âœ… IVF ë°ì´í„° ìƒê´€ê´€ê³„ ë¶„ì„
print("ğŸ“Œ IVF ë°ì´í„° ìƒê´€ê´€ê³„ ë¶„ì„ ê²°ê³¼")
df_ivf_corr = calculate_correlations(df_IVF)
print(df_ivf_corr)

# âœ… DI ë°ì´í„° ìƒê´€ê´€ê³„ ë¶„ì„
print("\nğŸ“Œ DI ë°ì´í„° ìƒê´€ê´€ê³„ ë¶„ì„ ê²°ê³¼")
df_di_corr = calculate_correlations(df_DI)
print(df_di_corr)
```
